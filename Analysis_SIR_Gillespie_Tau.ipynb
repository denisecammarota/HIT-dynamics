```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import lambertw
import seaborn as sns 
import pandas as pd
sns.set_style('ticks')
np.random.seed(123)
```

```python
def cases(beta,gamma,eps,I0):
    R0 = beta/gamma 
    alpha = (1/R0) + eps + I0
    beta2 = np.log((1/R0) + eps)
    gamma = R0
    C = (1/gamma)*(lambertw(-gamma*np.exp(beta2 - alpha*gamma)) + (alpha*gamma))
    return np.real(C)
```
