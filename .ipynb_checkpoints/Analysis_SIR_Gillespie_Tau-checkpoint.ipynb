```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
sns.set_style('ticks')
np.random.seed(123)
```

## First simulations $\epsilon = 0.01$

```python
fsize_eps1 = []
std_fsize_eps1 = []
ftime_eps1 = []
std_ftime_eps1 = []
```

```python
files = ['GTP_fsize_0.1_0.01_500000.csv']
N_vec = [1000,5000,10000,50000,100000,500000]
df = np.loadtxt(files[-1])
eps = 0.01
for i in range(len(N_vec)):
    fsize_eps1.append(np.mean(df[i]/N_vec[i]))
    std_fsize_eps1.append(np.std(df[i]/N_vec[i]))
    plt.title(r'$N =$ '+str(N_vec[i]))
    bins_vec = np.arange(0,0.1,0.0025)
    plt.hist(np.array(df[i])/N_vec[i], bins = bins_vec, edgecolor = 'black', color = 'red')
    plt.ylabel('Frequency')
    plt.xlabel('Outbreak Size (Propotion)')
    plt.axvline(eps, color = 'k')
    plt.show()
```

```python
files = ['GTP_ftime_0.1_0.01_1000.csv','GTP_ftime_0.1_0.01_5000.csv',
         'GTP_ftime_0.1_0.01_10000.csv','GTP_ftime_0.1_0.01_50000.csv',
        'GTP_ftime_0.1_0.01_100000.csv','GTP_ftime_0.1_0.01_500000.csv']
N_vec = [1000,5000,10000,50000,100000,500000]
df = np.loadtxt(files[-1])
eps = 0.01
for i in range(len(N_vec)):
    plt.title(r'$N =$ '+str(N_vec[i]))
    ftime_eps1.append(np.mean(df[i]))
    std_ftime_eps1.append(np.std(df[i]))
    #bins_vec = np.arange(0,0.1,0.0025)
    plt.hist(np.array(df[i]),  edgecolor = 'black', color = 'red')
    plt.ylabel('Frequency')
    plt.xlabel('Outbreak Size (Propotion)')
    plt.show()
```

### Plots as functions of $N$

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12,6))
ax1.errorbar(N_vec, fsize_eps1, std_fsize_eps1, fmt='-o', label = r'$\epsilon =$ 0.01')
ax1.set_xlabel(r'$N$ (log)', fontsize = 14)
ax1.set_xscale('log')
ax1.set_ylabel(r'$S_{\infty}$', fontsize = 14)
ax2.errorbar(N_vec, ftime_eps1, std_ftime_eps1, fmt='-o', label = r'$\epsilon =$ 0.01')
ax2.set_xlabel(r'$N$ (log)', fontsize = 14)
ax2.set_ylabel(r'$t_{\infty}$', fontsize = 14)
ax2.set_xscale('log')
ax1.legend()
ax2.legend()
```
